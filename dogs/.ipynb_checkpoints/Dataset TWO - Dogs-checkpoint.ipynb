## Make a new column called `monochrome` that is True for any animal that only has black, white or grey as one of its colors. How many animals are monochrome?

```python
#covid = df[df.covid_related == True]
merged['monochrome'] = (merged.animal_dominant_color.str.lower().isin(['black', 'white', 'grey', np.nan]))\
      & (merged.animal_secondary_color.str.lower().isin(['black','white','grey', np.nan]))\
      & (merged.animal_third_color.str.lower().isin(['black','white','grey', np.nan]))
                                              
merged.monochrome.value_counts()                                            
                                                                                            
                                              
```
